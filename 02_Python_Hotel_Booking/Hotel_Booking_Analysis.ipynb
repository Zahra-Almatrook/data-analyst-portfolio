```python
# Section 3 — Data Cleaning

# Step 1 — Check missing values before cleaning
print("=== MISSING VALUES BEFORE CLEANING ===")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Step 2 — Fix missing values

# Fill missing country with most common country
df['country'] = df['country'].fillna(df['country'].mode()[0])

# Fill missing agent with 0 (no agent)
df['agent'] = df['agent'].fillna(0)

# Fill missing company with 0 (no company)
df['company'] = df['company'].fillna(0)

# Drop children nulls — only 4 rows
df = df.dropna(subset=['children'])

# Verify no more missing values
print("=== MISSING VALUES AFTER CLEANING ===")
print(df.isnull().sum()[df.isnull().sum() > 0])
print("\nIf nothing printed above — no more missing values! ✅")
```

```python
# Step 3 — Fix data types

# Convert agent and company to integer
df['agent'] = df['agent'].astype(int)
df['company'] = df['company'].astype(int)
df['children'] = df['children'].astype(int)

print("Data types fixed successfully!")
print(df[['agent', 'company', 'children']].dtypes)
```

```python
# Step 4 — Remove invalid rows

# Remove rows where adults, children and babies are all 0
# These are invalid bookings with no guests
invalid = df[(df['adults'] == 0) & 
             (df['children'] == 0) & 
             (df['babies'] == 0)]
print(f"Invalid bookings with no guests: {len(invalid)}")
df = df[~((df['adults'] == 0) & 
          (df['children'] == 0) & 
          (df['babies'] == 0))]

# Remove rows where ADR is negative or extremely high
print(f"Rows with negative ADR: {len(df[df['adr'] < 0])}")
df = df[df['adr'] >= 0]
df = df[df['adr'] < 5000]

print(f"\nDataset shape after cleaning: {df.shape}")
```

```python
# Step 5 — Create new useful columns

# Total guests column
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Total nights column
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Total revenue column (ADR x total nights)
df['total_revenue'] = df['adr'] * df['total_nights']

# Arrival date column
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d'
)

# Season column
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['arrival_date'].dt.month.apply(get_season)

print("New columns created successfully!")
print(df[['total_guests', 'total_nights', 
          'total_revenue', 'arrival_date', 'season']].head())
```

```python
# Step 6 — Final verification
print("=== FINAL DATASET SUMMARY ===")
print(f"Total Rows: {df.shape[0]:,}")
print(f"Total Columns: {df.shape[1]}")
print(f"\nDate Range: {df['arrival_date'].min()} to {df['arrival_date'].max()}")
print(f"Total Revenue: ${df['total_revenue'].sum():,.2f}")
print(f"Average Daily Rate: ${df['adr'].mean():,.2f}")
print(f"Cancellation Rate: {df['is_canceled'].mean()*100:.1f}%")
```

```python
# Section 4 — Exploratory Data Analysis

# Overall booking summary
print("=== OVERALL BOOKING SUMMARY ===")
print(f"Total Bookings: {len(df):,}")
print(f"Total Cancelled: {df['is_canceled'].sum():,}")
print(f"Total Completed: {(df['is_canceled']==0).sum():,}")
print(f"Cancellation Rate: {df['is_canceled'].mean()*100:.1f}%")
print(f"\nCity Hotel Bookings: {len(df[df['hotel']=='City Hotel']):,}")
print(f"Resort Hotel Bookings: {len(df[df['hotel']=='Resort Hotel']):,}")
```

```python
# Cancellation analysis by hotel type
print("=== CANCELLATION BY HOTEL TYPE ===")
cancel_hotel = df.groupby('hotel')['is_canceled'].agg(
    ['sum', 'count', 'mean'])
cancel_hotel.columns = ['Cancelled', 'Total', 'Cancellation Rate']
cancel_hotel['Cancellation Rate'] = (
    cancel_hotel['Cancellation Rate'] * 100).round(1)
print(cancel_hotel)

print("\n=== CANCELLATION BY MARKET SEGMENT ===")
cancel_segment = df.groupby('market_segment')['is_canceled'].agg(
    ['sum', 'count', 'mean'])
cancel_segment.columns = ['Cancelled', 'Total', 'Cancellation Rate']
cancel_segment['Cancellation Rate'] = (
    cancel_segment['Cancellation Rate'] * 100).round(1)
print(cancel_segment.sort_values('Cancellation Rate', ascending=False))
```

```python
# Section 6 — Key Insights & Conclusions

print("=" * 60)
print("   KEY INSIGHTS — HOTEL BOOKING ANALYSIS")
print("=" * 60)

# Insight 1 — Overall cancellation
print("\n📊 INSIGHT 1 — CANCELLATION OVERVIEW")
print(f"   Total Bookings: {len(df):,}")
print(f"   Total Cancelled: {df['is_canceled'].sum():,}")
print(f"   Overall Cancellation Rate: {df['is_canceled'].mean()*100:.1f}%")

# Insight 2 — Hotel type cancellation
print("\n🏨 INSIGHT 2 — CANCELLATION BY HOTEL TYPE")
for hotel in df['hotel'].unique():
    rate = df[df['hotel']==hotel]['is_canceled'].mean()*100
    total = len(df[df['hotel']==hotel])
    print(f"   {hotel}: {rate:.1f}% cancellation rate ({total:,} bookings)")

# Insight 3 — Busiest month
print("\n📅 INSIGHT 3 — BUSIEST BOOKING MONTH")
month_order = ['January', 'February', 'March', 'April',
               'May', 'June', 'July', 'August',
               'September', 'October', 'November', 'December']
busiest = df['arrival_date_month'].value_counts().index[0]
busiest_count = df['arrival_date_month'].value_counts().values[0]
quietest = df['arrival_date_month'].value_counts().index[-1]
quietest_count = df['arrival_date_month'].value_counts().values[-1]
print(f"   Busiest Month: {busiest} ({busiest_count:,} bookings)")
print(f"   Quietest Month: {quietest} ({quietest_count:,} bookings)")

# Insight 4 — Revenue
print("\n💰 INSIGHT 4 — REVENUE OVERVIEW")
print(f"   Total Revenue: ${df['total_revenue'].sum():,.2f}")
print(f"   Average Daily Rate: ${df['adr'].mean():,.2f}")
city_adr = df[df['hotel']=='City Hotel']['adr'].mean()
resort_adr = df[df['hotel']=='Resort Hotel']['adr'].mean()
print(f"   City Hotel Avg ADR: ${city_adr:.2f}")
print(f"   Resort Hotel Avg ADR: ${resort_adr:.2f}")

# Insight 5 — Top country
print("\n🌍 INSIGHT 5 — TOP BOOKING COUNTRIES")
top3 = df['country'].value_counts().head(3)
for country, count in top3.items():
    pct = count/len(df)*100
    print(f"   {country}: {count:,} bookings ({pct:.1f}%)")

# Insight 6 — Lead time
print("\n⏰ INSIGHT 6 — LEAD TIME & CANCELLATIONS")
cancelled_lead = df[df['is_canceled']==1]['lead_time'].mean()
not_cancelled_lead = df[df['is_canceled']==0]['lead_time'].mean()
print(f"   Avg Lead Time (Cancelled): {cancelled_lead:.0f} days")
print(f"   Avg Lead Time (Not Cancelled): {not_cancelled_lead:.0f} days")
print(f"   Cancelled bookings are made {cancelled_lead-not_cancelled_lead:.0f} days earlier on average")

# Insight 7 — Season
print("\n🌤️ INSIGHT 7 — SEASONAL TRENDS")
season_bookings = df.groupby('season').size().sort_values(ascending=False)
for season, count in season_bookings.items():
    pct = count/len(df)*100
    print(f"   {season}: {count:,} bookings ({pct:.1f}%)")

# Insight 8 — Market segment
print("\n🎯 INSIGHT 8 — HIGHEST CANCELLATION SEGMENT")
highest_cancel = df.groupby('market_segment')['is_canceled'].mean()*100
highest = highest_cancel.idxmax()
highest_rate = highest_cancel.max()
lowest = highest_cancel.idxmin()
lowest_rate = highest_cancel.min()
print(f"   Highest Cancellation: {highest} ({highest_rate:.1f}%)")
print(f"   Lowest Cancellation: {lowest} ({lowest_rate:.1f}%)")

# Insight 9 — Stay duration
print("\n🛏️ INSIGHT 9 — STAY DURATION")
print(f"   Average Nights per Stay: {df['total_nights'].mean():.1f} nights")
city_nights = df[df['hotel']=='City Hotel']['total_nights'].mean()
resort_nights = df[df['hotel']=='Resort Hotel']['total_nights'].mean()
print(f"   City Hotel Avg Stay: {city_nights:.1f} nights")
print(f"   Resort Hotel Avg Stay: {resort_nights:.1f} nights")

# Insight 10 — Deposit type
print("\n💳 INSIGHT 10 — DEPOSIT TYPE IMPACT")
deposit_cancel = df.groupby('deposit_type')['is_canceled'].mean()*100
for deposit, rate in deposit_cancel.items():
    print(f"   {deposit}: {rate:.1f}% cancellation rate")

print("\n" + "=" * 60)
print("   ANALYSIS COMPLETE!")
print("=" * 60)
```
